```python
# Fazendo da primeira forma estava demorando cerca de 45 segundo para rodar
num = 1000000
max_cont = 0
max_num = 0
i = 0

# Vamos pesquisar por todos os valores
while i < num:
    
    j = i
    cont = 0
    
    # Enquanto não chegarmos ao 1, vamos aplicar a condição
    while j > 1:        
        if(j%2 == 0):
            j = j/2
            cont += 1
        else: 
            j = 3*j+1
            cont += 1
            
    # Se possui mais números na cadeia, vamos considerar esse valor     
    if(cont > max_cont):
        max_cont = cont
        max_num = i
        
    i += 1
    
print(max_num, max_cont)
```

```python
# Refazendo rodou bem mais rápido

# Vamos refazer usando um dicionário, pois se achamos que o número 13 possui 10 elementos na cadeia,
# quando chegarmos nesse valor sabemos que serão os mesmos 10 elementos
dicionario = {}
num = 1000000
max_cont = 0
max_num = 0
i = 2

while i < num:
    
    j = i
    cont = 1
    while j > 1:        
        if(j%2 == 0):
            j = j/2
            cont += 1
        else: 
            j = 3*j+1
            cont += 1
            
        if(j in dicionario):
            # Usando os valores que já estão no dicionário
            cont += (dicionario[j] - 1)
            # E finalizando o while
            j = 1

    dicionario[i] = cont
    
    if(cont > max_cont):
        max_cont = cont
        max_num = i
        
    i += 1
    
print(max_num, max_cont)
```
