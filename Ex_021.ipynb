```python
# Vamos buscar os divisores de todos os nmeros at o limite
limite = 10000

# E vamos criar um dicionrio com esses valores
dicionario_divisores = {}

i = 0
while i < limite:
    
    j = 2 
    soma_divisores = 1
    
    while j <= i/2:
        # Verificando se o nmero  divisor de todos os elementos da lista
        if i % j == 0:
            soma_divisores += j

        j += 1

    dicionario_divisores[i] = soma_divisores
    
    i += 1
    
print(dicionario_divisores)   
```

```python
# Depois vamos percorrer todo esse dicionrio validando se d(a) = b and d(b) = a
i = 0

par_amigavel = []

while i < limite:
    a = i
    b = dicionario_divisores[i]
    
    # Vamos fazer todas as condies mas tambm validar se no estamos fora da lista
    if (b < limite and dicionario_divisores[a] == b and dicionario_divisores[b] == a and a != b):
        
        # Se ainda no temos esse par na nossa lista, vamos adicionar
        if a not in par_amigavel:
            par_amigavel.append(a)
            par_amigavel.append(b)
            
    i += 1
```

```python
# Somando os valores da lista

soma = 0
for i in par_amigavel:
    soma += i
    
print(soma)
```
